```python
# Nacitanie datoveho suboru treningovej alebo testovacej mnoziny na preklad
import pandas as pd

df = pd.read_csv("test_df.csv")
```

```python
# Definicia funkcie na preklad textu z ceskeho do anglickeho jazyka
from deep_translator import GoogleTranslator

def translate_row(row):
    translated_text = GoogleTranslator(source="cs", target="en").translate(row)
    return translated_text
```

```python
df["translated_text"] = df["title_with_description"].apply(translate_row)
```

```python
# Ukazka zaznamov so vsetkymi atributmi, vratane prelozeneho textu v anglickom jazyku
df
```
