# Informácie o kódoch CPV tréningovej a testovacej množiny dát

## Tréningová množina dát

```python
# Nacitanie treningovej mnoziny zaznamov
df1 = pd.read_csv("preklad_train_final.csv")
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Vykreslenie stlpcoveho diagramu poctu treningovych prikladov pre kazdy oddiel schemy CPV
def plot_number_of_records(df, text_column, label_column):
	average_character_count_per_label = df.groupby(df[label_column].map(lambda x: x[:2]))[text_column].size()

	plt.figure(figsize=(10, 3))
	average_character_count_per_label.plot(kind="bar", color="steelblue")

	for i, loss in enumerate(average_character_count_per_label):
		plt.text(i, loss + 5, f"{loss}", ha="center", va="bottom", fontsize="x-small")
	plt.xlabel("Oddiel CPV (na prvej úrovni hierarchie)")
	plt.xticks(rotation=90)
	plt.ylabel("Záznamy v testovacej množine")

	plt.legend(["Počet záznamov"])
	plt.grid(True, axis="y", alpha=0.3)

	plt.show()

plot_number_of_records(df1, "cs_title_with_description", "cpv_level1")
```

```python
# Definicia funkcie pre vykreslenie stlpcoveho diagramu priemerneho poctu znakov treningovych zaznamov pre kazdy oddiel schemy CPV
def plot_average_number_characters(df, text_column, label_column):
	average_character_count_per_label = df.groupby(df[label_column].map(lambda x: x[:2]))[text_column].apply(lambda x: x.str.len().mean())

	plt.figure(figsize=(10, 3))
	average_character_count_per_label.plot(kind="bar", color="steelblue")
	plt.xlabel("Oddiel CPV (na prvej úrovni hierarchie)")
	plt.xticks(rotation=90)
	plt.ylabel("Priemerný počet znakov")

	plt.legend(["Počet znakov"])
	plt.grid(True, axis="y", alpha=0.3)

	plt.show()
```

```python
# Vykreslenie stlpcoveho diagramu priemerneho poctu znakov treningovych zaznamov v ceskom jazyku pre kazdy oddiel
plot_average_number_characters(df1, "cs_title_with_description", "cpv_level1")
```

```python
# Vykreslenie stlpcoveho diagramu priemerneho poctu znakov treningovych zaznamov v anglickom jazyku pre kazdy oddiel
plot_average_number_characters(df1, "en_title_with_description", "cpv_level1")
```
