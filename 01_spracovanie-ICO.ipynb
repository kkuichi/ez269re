```python
import requests, time

# dokumentacia k API rozhraniu - Swagger UI: [https://ares.gov.cz/swagger-ui/#/ekonomicke-subjekty/vratEkonomickySubjekt]
# funkcia na ziskanie udajov z HTTP GET poziadavky na cesky Administrativny register ekonomickych subjektov
def get_data_from_api(ico):
    time.sleep(1)
    url = f"https://ares.gov.cz/ekonomicke-subjekty-v-be/rest/ekonomicke-subjekty/{ico}"
    response = requests.get(url)
    if response.status_code == 200:
        json_data = response.json()
        obchodni_jmeno = json_data["obchodniJmeno"]
        if "sidlo" in json_data and "nazevObce" in json_data["sidlo"]:
          nazev_obce = json_data["sidlo"]["nazevObce"]
        else:
          nazev_obce = None
        pravni_forma = json_data["pravniForma"]
        if nazev_obce is None:
          print(response.status_code, " ", ico, " ----- ", nazev_obce)
          return obchodni_jmeno, pravni_forma, nazev_obce
        print(response.status_code, " ", ico, " ", nazev_obce)
        return obchodni_jmeno, pravni_forma, nazev_obce
    else:
        return None, None, None

df_gov["obchodni_jmeno"], df_gov["pravni_forma"], df_gov["nazev_obce"] = zip(*df_gov["ipo"].apply(lambda x: get_data_from_api(x)))

# Ulozenie identifikatorov podnikatelskych subjektov do CSV suboru
df_gov.to_csv("ico.csv", index=False)
```
